### Coherent and Uncoherent band attachments
A link can be transformed into a different link by moves called band attachments. Band attachments can be either coherent or uncoherent. Coherent band attachments change the number of components of the link by 1, while uncoherent leave it unchanged.

<img src="band.png" width=300 height=600 />

On the top, a local picture for a coherent band attachment, on the bottom, for an uncoherent one.

```python
GG =  load_knot('4_1')
draw_grid(GG, markings='XO')
B = coherent_bs(GG, 3, 'c')
draw_grid(B, markings='XO')
print('The diagram has', number_of_components(GG), 'connected component.')
print('After an coherent band attachment, it has', number_of_components(B), 'components.')
```

```python
GG =  load_knot('5_1')
draw_grid(GG, markings='XO')
K = uncoherent_bs(GG,3)
draw_grid(K, markings='XO')
print('The diagram has', number_of_components(GG), 'connected component.')
print('After an uncoherent band attachement, it still has', number_of_components(K), 'component.')
```

```python
draw_grid(load_knot('3_1'), markings='XO')
print('Link diagrams differing from GG by a single coherent band attachement:')
for i in range(4):
    GG = load_knot('3_1')
    draw_grid(coherent_bs(GG, i, 'r'), markings='XO')
    print('Band attachement performed at the row N {}'.format(i))
    GG = load_knot('3_1')
    draw_grid(coherent_bs(GG, i, 'c'), markings='XO')
    print('Band attachement performed at the column N {}'.format(i))
```

```python
draw_grid(load_knot('3_1'), markings='XO')
print('Knot diagrams differing from GG by a single uncoherent band attachement:')
for i in range(4):
    GG = load_knot('3_1')
    draw_grid(uncoherent_bs(GG, i, 'r'), markings='XO')
    print('Band attachement performed at the row N {}'.format(i))
    GG = load_knot('3_1')
    draw_grid(uncoherent_bs(GG, i, 'c'), markings='XO')
    print('Band attachement performed at the column N {}'.format(i))
```

## Crossing changes as interleaving commutations

Commuting two rows or columns that are interleaving as the effect of performing a crossing change on the underlying link diagram. 

```python
GG = load_knot('5_2')
CC = commute_columns(GG, 3,interleaving='Y')
draw_grid(GG, markings='XO')
draw_grid(CC, markings='XO')
```

```python
GG = load_knot('6_2')
CC = commute_rows(GG, 2,interleaving='Y')
draw_grid(GG, markings='XO')
draw_grid(CC, markings='XO')
```
